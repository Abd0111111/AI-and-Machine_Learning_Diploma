```python
import pymongo
from pymongo import MongoClient
from bson.objectid import ObjectId
```

```python
# في كزا package بستخدمها علشان ابني ال API
from flask import Flask , request ,jsonify
import threading
import time
```

```python
app = Flask(__name__)


client = MongoClient("mongodb://localhost:27017/")


db = client['CustomersDB']


customers_col = db['Customers']
```

```python
@app.route('/customers/<id>', methods=['DELETE'])
# Define an API endpoint at /customers/<id>
# This endpoint responds only to DELETE requests
# DELETE is used to remove data

def delete_customer(id):
    # Define a function that runs when the endpoint is called
    # id is received from the URL

    try:
        obj_id = ObjectId(id)
        # Convert the id from string to MongoDB ObjectId
        # MongoDB requires ObjectId type to match _id field

    except:
        # This runs if the id format is invalid
        return jsonify({"message": "Invalid customer id"}), 400
        # Return error response if id is not valid

    result = customers_col.delete_one({"_id": obj_id})
    # Delete one document that matches the given _id

    if result.deleted_count == 1:
        # Check if a document was actually deleted
        return jsonify({"message": "Customer deleted successfully"})
        # Send success response

    else:
        # This runs if no document matched the id
        return jsonify({"message": "Customer not found"}), 404
        # Send not found response
```

```python
if __name__ == "__main__":
    app.run(debug=True)
```
