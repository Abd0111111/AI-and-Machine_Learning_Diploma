```python
import pymongo
from pymongo import MongoClient
from bson.objectid import ObjectId
```

```python
# في كزا package بستخدمها علشان ابني ال API
from flask import Flask , request ,jsonify
import threading
import time
```

```python
app = Flask(__name__)
# Create a Flask application instance
# This object represents your API server
# It handles requests and responses

client = MongoClient("mongodb://localhost:27017")
# Create a connection to MongoDB
# localhost means the database is running on the same machine
# 27017 is the default MongoDB port

db = client['CustomersDB']
# Select the database named CustomersDB
# If it does not exist MongoDB will create it automatically

customers_col = db['Customers']
# Select the collection named customers
# A collection is similar to a table in SQL
# This is where customer data will be stored
```

```python
@app.route('/customers', methods=['GET'])
# Define an API endpoint at /customers
# This endpoint responds only to GET requests
# GET is used to read data

def get_customers():
    # Define a function that runs when the endpoint is called

    customers = []
    # Create an empty list
    # This list will hold all customer records

    for doc in customers_col.find():
        # Loop through all documents in the customers collection
        # find() returns all records from MongoDB

        doc['_id'] = str(doc['_id'])
        # Convert MongoDB ObjectId to string
        # ObjectId cannot be returned as JSON directly

        customers.append(doc)
        # Add the modified document to the customers list
        # This prepares the data for the response

    return jsonify(customers)
    # Convert the list to JSON format
    # Send the JSON response back to the client
```

```python
def run_flask():
    app.run(port=5000, debug=False, use_reloader=False)

threading.Thread(target=run_flask).start()
time.sleep(1)
```
