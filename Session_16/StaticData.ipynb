```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # Basic backage
import seaborn as sns
```

```python

days = ["Sat", "Sun", "Mon", "Tue", "Wed", "Thu", "Fri"]
cairo_temps = [30, 32, 31, 29, 28, 27, 26]
alex_temps = [25, 26, 27, 28, 28, 29, 30]
sharm_temps = [33, 33, 34, 30, 29, 28, 27]
```

```python
# 1. create figure (canvas)
```

```python
plt.figure(figssize = (8,5))

```

```python
    # Draw the plot line
```

```python
plt.plot(days, cairo_temps, label = "Cairo", color = "orange", linewidth = 2, linestyle = "--")
plt.plot(days, sharm_temps, label = "Sharm", color = "red", linewidth = 2, linestyle = "-")
plt.plot(days, alex_temps, label = "Alex", color = "blue", linewidth = 2, linestyle = "-")

plt.title("weekly temprature comparison between cities ")
plt.xlabel("days of weeks")
plt.ylabel(" temprature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Heat map
```

```python
numiric_curr = tips.corr(numeric_only = True)
plt.figure(figsize = (8,5))
sns.heatmap(numiric_curr.corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.show()
```
