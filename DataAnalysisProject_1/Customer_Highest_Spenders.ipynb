```python
import pandas as pd
from pymongo import MongoClient
from bson.objectid import ObjectId

```

```python
# define connection string
Orders = MongoClient('mongodb://localhost:27017')
```

```python
db = Orders["Retail_Business"]
```

```python
orders_col = db["Orders"]
```

```python
# 2) Read data from Orders collection
data = list(orders_col.find())
df = pd.DataFrame(data)

print("Number of records:", df.shape[0])
df.head(10)
```

```python
# Extract useful fields from nested documents

df["customer_id"] = df["customer"].apply(lambda x: x["id"])
df["customer_name"] = df["customer"].apply(lambda x: x["name"])

df["branch_id"] = df["branch"].apply(lambda x: x["id"])
df["branch_name"] = df["branch"].apply(lambda x: x["name"])

df["sale_id"] = df["sale"].apply(lambda x: x["id"])
df["sale_date"] = pd.to_datetime(df["sale"].apply(lambda x: x["date"]))
df["sale_total"] = df["sale"].apply(lambda x: x["total_amount"])

df["product_id"] = df["product"].apply(lambda x: x["id"])
df["product_name"] = df["product"].apply(lambda x: x["name"])
df["product_category"] = df["product"].apply(lambda x: x["category"])
df["product_price"] = df["product"].apply(lambda x: x["price"])
df["product_quantity"] = df["product"].apply(lambda x: x["quantity"])

# عرض أول صفوف بعد الفرد
df[[
    "customer_id", "customer_name",
    "branch_name",
    "sale_total", "sale_date",
    "product_name", "product_quantity"
]].head()
```

```python
# STEP 2 — Top Customers Analysis

# أولًا: نجيب Orders بدون تكرار (حسب sale_id + customer)
orders_unique = df[["sale_id", "customer_id", "customer_name", "sale_total"]].drop_duplicates() # ديه علشان تشبل اي تكرار

# نعمل Groupby
top_customers = (
    orders_unique
    .groupby(["customer_id", "customer_name"])
    .agg(
        total_spent=("sale_total", "sum"),
        total_orders=("sale_id", "count")
    )
    .reset_index()
    .sort_values(by="total_spent", ascending=False)
)

top_customers.head(20)
```
