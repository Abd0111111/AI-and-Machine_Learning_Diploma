```python
import pymongo
from pymongo import MongoClient
from bson.objectid import ObjectId
```

```python
# define connection string
client = MongoClient('mongodb://localhost:27017')
```

```python
db = client["CustomersDB"]
```

```python
customers_col = db["Customers"]
```

```python
# fun to update a customer by id
# This function updates a customer document in the customers collection
def update_customer(customer_id, customer):

    try:
        # Attempt to update one document that matches the given _id
        result = customers_col.update_one(
            {"_id": ObjectId(customer_id)},   # Filter by ObjectId
            {"$set": {
                "Name": customer['Name'],
                "Age": customer['Age'],
                "Gender": customer['Gender'],
                "Region": customer['Region']
            }}
        )

        # Check if a document was actually updated
        if result.matched_count == 0:
            return False, "Customer not found"

        elif result.modified_count == 0:
            return True, "Customer found but data was already up-to-date"

        else:
            return True, "Customer updated successfully"

    except Exception as e:
        return False, str(e)
```

```python
# Example usage of update_customer function
updated_customer_info = {
    "Name": "Alice Johnson",
    "Age": 29,
    "Gender": "Female",
    "Region": "Giza"
}

# Call the function and receive the result
success, message = update_customer("69625ad38ba5c5fafabe7baa", updated_customer_info)

# Check if the update operation was successful
if success:
    print(message)
else:
    print("Error updating customer:", message)
```
